```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Load the data from the path using pandas
data = pd.read_csv('C:/Users/augus/OneDrive/DTU/Fagprojekt/wetransfer_main-m_2024-02-21_1730/df.csv')

# Print the first 5 rows of the dataframe
print(data.head())

# Dimensions of the dataframe
print(data.shape)
```

```python
# Find NaN values in the dataframe
print(data.isna().sum())
```

```python
# Count unique mouse IDs
print(data['mouseID'].unique())
```

```python
# Print the mouse with ID -1
print(data[data['mouseID'] == -1])
```

```python
# Print header in latex format
print(data.head().to_latex())
```

```python
# Compute hypnogram of the sleep stages over the entire recording, epoch by epoch

# Create a new dataframe with the sleep stages
sleep_stages = data[['epoch', 'sleepstage']]
print(sleep_stages.head())

# Count the number of epochs for each sleep stage
print(sleep_stages['sleepstage'].value_counts())

# Plot the sleep stages
plt.plot(sleep_stages['sleepstage'][:100])
plt.xlabel('Epoch')
plt.ylabel('Sleep stage')
plt.title('Hypnogram')
plt.show()

```

```python
# Extract mouse id 1 from lab 1
mouse1 = data[(data['mouseID'] == 1) & (data['lab'] == 1)]
print(mouse1.head())
# 1 = REM, 2 = Wake,  3 = NREM

# Make Wake = 3, NREM = 2, REM = 1
mouse1['sleepstage'] = mouse1['sleepstage'].replace(2, 31)
mouse1['sleepstage'] = mouse1['sleepstage'].replace(3, 2)
mouse1['sleepstage'] = mouse1['sleepstage'].replace(31, 3)


# Count the number of epochs for each sleep stage
print(mouse1['sleepstage'].value_counts())

# Lentgh of the recording
print(mouse1.shape)


# Plot the sleep stages for mouse 1
plt.plot(mouse1['sleepstage'][300:500])
plt.xlabel('Epoch')
plt.ylabel('Sleep stage')
plt.title('Hypnogram for mouse 1')
plt.show()
```

```python
# Mouse 2, lab 1
mouse2 = data[(data['mouseID'] == 2) & (data['lab'] == 1)]
print(mouse2[:15])
```

```python
# Histrogram of the sleep stages for mouse 2
plt.hist(mouse2['sleepstage'])
plt.xlabel('Sleep stage')
plt.ylabel('Count')
# X-axis labels
plt.xticks([1.1, 2.1, 2.9], ['REM', 'NREM', 'Wake']) # 1 = REM, 2 = NREM,  3 = Wake


plt.title('Histogram of sleep stages for mouse 2')
plt.show()
```
